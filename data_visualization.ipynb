```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
fig , ax = plt.subplots( 1,2 , figsize = (12,6))
sns.scatterplot(data = df , x = 'petal_length' , y = 'petal_width' , hue = 'species' ,ax = ax[0])
sns.scatterplot(data = df , x = 'sepal_length' , y = 'sepal_width' , hue = 'species' , ax = ax[1])

ax[0].set_xlabel('petal length (cm)')
ax[0].set_ylabel('petal width (cm)')
ax[0].set_title('species : petal length vs petal width')
ax[0].legend(loc = 'upper left')

ax[1].set_xlabel('sepal length (cm)')
ax[1].set_ylabel('sepal width (cm)')
ax[1].set_title('species : sepal length vs sepal width')
ax[1].legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

sns.histplot(data=df, x='sepal_length' , ax=ax[0, 0])
ax[0, 0].set_title('Sepal Length Distribution')

sns.histplot(data=df, x='sepal_width', ax=ax[0, 1])
ax[0, 1].set_title('Sepal Width Distribution')

sns.histplot(data=df, x='petal_length', ax=ax[1, 0])
ax[1, 0].set_title('Petal Length Distribution')

sns.histplot(data=df, x='petal_width', ax=ax[1, 1])
ax[1, 1].set_title('Petal Width Distribution')

fig.suptitle('Distribution of Iris Features', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

sns.boxplot(data=df, x='species', y='sepal_length', ax=ax[0, 0])
sns.boxplot(data=df, x='species', y='sepal_width', ax=ax[0, 1])
sns.boxplot(data=df, x='species', y='petal_length', ax=ax[1, 0])
sns.boxplot(data=df, x='species', y='petal_width', ax=ax[1, 1])

ax[0, 0].set_title('Sepal Length by Species')
ax[0, 1].set_title('Sepal Width by Species')
ax[1, 0].set_title('Petal Length by Species')
ax[1, 1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```
